### Defi llama TVL 

```python
import requests
```

```python
r = requests.get('https://api.llama.fi/protocols')
```

```python
df_protocol_tvl = pd.json_normalize(r.json())
```

```python
df_protocol_tvl_eth = df_protocol_tvl[['name','address', 'symbol', 'gecko_id', 'url' , 'mcap', 'tvl','chainTvls.Ethereum', 'chain', 'category']]
df_protocol_tvl_eth.dropna(subset='address', inplace=True)
```

```python
df_protocol_tvl_eth['relative_share_of_tvl_on_eth'] = df_protocol_tvl_eth['chainTvls.Ethereum']/ df_protocol_tvl_eth['tvl']
```

```python
df_protocol_tvl_eth_f = df_protocol_tvl_eth[(((df_protocol_tvl['chainTvls.Ethereum'] >= 300_000_000) | (df_protocol_tvl['mcap'] >= 200_000_000 ))) #mcap 200MM and 300MM TVL
    & ((df_protocol_tvl_eth['address'].apply(lambda x: len(x) != 42)) != True) #ERC20 Compliant - address longer than 42 char
    & (df_protocol_tvl_eth['relative_share_of_tvl_on_eth'] > 0.5)] # the majority of the tvl is on ethereum
```

```python
df_protocol_tvl_eth_f.drop_duplicates(subset='address', inplace=True)
```

```python
### merge with eth govt tokens according to coingecko 
df_govt_token_f = df_govt_token[['name', 'platforms.ethereum', 'symbol', 'id', 'links.homepage', 'market_cap']]
df_govt_token_f['tvl'] = 0
df_govt_token_f['chainTvls.Ethereum'] = 0
df_govt_token_f['chain'] = 'Ethereum'
df_govt_token_f['category'] = 'Governance'

df_govt_token_f.columns = ['name','address', 'symbol', 'gecko_id', 'url' , 'mcap', 'tvl','chainTvls.Ethereum', 'chain', 'category']
```

```python
df_final_tokens = pd.concat([df_protocol_tvl_eth_f, df_govt_token_f])
```

```python
df_final_tokens.symbol = df_final_tokens.symbol.str.upper() 
```

```python
df_final_tokens.drop_duplicates(subset='symbol', inplace=True)
```

```python
### manual refinement 

remove = [
    'CRO',   # utility token of cronos chain - no govt 
    'SNX',   # see Nadler & Schaer 2020
    'WBTC',  # Wrapped asset - no govt right
    'HBTC',  # Wrapped asset - no govt right
    'CBETH', # Staked asset - no govt right
]
```

```python
df_final_tokens_new = df_final_tokens[df_final_tokens.symbol.isin(remove) != True]


```

```python
from web3 import Web3

df_final_tokens_new["address"] = df_final_tokens_new["address"].apply(lambda address: Web3.toChecksumAddress(address))
```

```python
df_final_tokens_new
```

```python
df_final_tokens_new.to_csv(join('./assets', 'df_final_token_selection_20221129.csv'))
```
